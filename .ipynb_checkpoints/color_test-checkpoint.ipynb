```python
from superminddpm import DDPM, DummyEpsModel
import torch
import matplotlib.pyplot as plt
from torchvision.utils import make_grid
from utils import plot_samples
from PIL import Image
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = DDPM(eps_model=DummyEpsModel(3), betas=(1e-4, 0.02), n_T=1000)
model.load_state_dict(torch.load("./contents/colors2/ddpm_mnist_colors.pth", map_location=device))
model.to(device)
model.eval();
```

```python
size = (3, 28, 28)
n = 2
with torch.no_grad():
    samples = model.sample(n, size, device)
```

```python
plot_samples(samples, normalize=True)
```

## Sampling process

```python
n = 2
x_t = torch.randn(n, *(3, 28, 28)).to(device)
torch.set_grad_enabled(False)
store_every= 10
stored = torch.Tensor(*(0, n, 3, 28, 28)).to(device)
stored_timestep = []
for t in range(1000, 0, -1):
    z = torch.randn(n, *(3, 28, 28)).to(device) if t > 1 else 0
    eps = model.eps_model(x_t, 0)
    x_t = (model.oneover_sqrta[t] * (x_t - eps * model.mab_over_sqrtmab[t]) + model.sqrt_beta_t[t] * z)
    if t%store_every == 0 or t==1:
        stored = torch.cat([stored, x_t[None, ...]], dim=0)
        stored_timestep.append(t)
```

### Every hundred

```python
subset = stored[::10]
subset_t = stored_timestep[::10]
for subtensor, t in zip(subset, subset_t):
    plot_samples(subtensor, title=f"t={t}", normalize=True)
```

```python
subset = stored[-10:]
subset_t = stored_timestep[-10:]
for subtensor, t in zip(subset, subset_t):
    plot_samples(subtensor, title=f"t={t}", normalize=True)
```
