```python
from superminddpm import DDPM, DummyEpsModel
import torch
import matplotlib.pyplot as plt
from torchvision.utils import make_grid
from PIL import Image
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = DDPM(eps_model=DummyEpsModel(3), betas=(1e-4, 0.02), n_T=1000)
model.load_state_dict(torch.load("./contents/colors2/ddpm_mnist_colors.pth", map_location=device))
model.to(device)
model.eval();
```

```python
def plot_samples(samples, title=None):
    n = samples.shape[0]
    fig, axs = plt.subplots(1, n, figsize=(n * 3, 3))
    if title:
        fig.suptitle(title, fontsize=20)
    for i in range(n):
        curr_axs = axs if n == 1 else axs[i]
        curr_axs.imshow(samples[i].permute(1, 2, 0).cpu())
    plt.show()
```

```python
size = (3, 28, 28)
n = 16
with torch.no_grad():
    samples = model.sample(n, size, device)
```
